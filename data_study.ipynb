```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("mteb/banking77")

train_data = dataset['train']
test_data = dataset['test']
```

```python
df = pd.DataFrame(train_data)
```

```python
label_counts = df["label_text"].value_counts()
percentages = label_counts / label_counts.sum() * 100

plt.figure(figsize=(10, 20))
plt.barh(label_counts.index, label_counts.values, color='skyblue')

# Add percentages at the end of each bar
for i, (count, pct) in enumerate(zip(label_counts.values, percentages)):
    plt.text(count + 0.5, i, f"{pct:.1f}%", va='center')

plt.xlabel('Count')
plt.title('Label Distribution (Counts + Percentages)')
plt.gca().invert_yaxis()  # largest bar on top
plt.show()
```
